```python
from math import e
def sigmoid(z):
  return 1/(1+e**-z)
```

```python
print(sigmoid(-10))
print(sigmoid(-1))
print(sigmoid(0))
print(sigmoid(1))
print(sigmoid(10))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(z):
  return 1.0/(1.0+np.exp(-z))
values=np.arange(-10,10,0.1) 
plt.plot(values, sigmoid(values))
plt.xlabel('z')
plt.ylabel('sigmoid(z)')
plt.title('sigmoid function')
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
def tanh(z):
  return (np.exp(z)-np.exp(-z))/(np.exp(z)+np.exp(-z))
values=np.arange(-10,10,0.1) 
plt.plot(values, tanh(values))
plt.xlabel('z')
plt.ylabel('tanh(z)')
plt.title('Tanh function')
plt.show
```
